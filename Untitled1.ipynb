```python
import math
a = [1,2,3]
b = [2,4,1]

def dot_product(v1,v2):
    return sum(x*y for x,y in zip(v1,v2))

def cosine_similarity(v1,v2):
    return dot_product(v1,v2)/(math.sqrt(dot_product(v1,v1))*math.sqrt(dot_product(v2,v2)))

print("Dot product:", dot_product(a,b))
print("Cosine similarity:", cosine_similarity(a,b))
```

problem 7


```python
scores = [2.0,1.0,0.1]
def softmax(scores):
    exp_scores = [math.exp(s) for s in scores]
    total = sum(exp_scores)
    return [e/total for e in exp_scores]

probs = softmax(scores)
print("Softmax probabilities:", probs)
print("Sum:", sum(probs))
```

```python
w = [0.2,-0.1]
b = 0.0
x = [1.0,2.0]
y_true = 1
lr = 0.5

y_hat = 1 if w[0]*x[0]+w[1]*x[1]+b >=0 else 0
w[0] += lr*(y_true - y_hat)*x[0]
w[1] += lr*(y_true - y_hat)*x[1]
b += lr*(y_true - y_hat)
print("Prediction:", y_hat)
print("Updated w:", w)
print("Updated b:", b)
```

problem 9

```python
X = [[1,1],[2,1],[2,3],[3,3]]
y = [0,0,1,1]
w = [0,0]; b = 0; lr = 0.1

for _ in range(10):
    for x_i, yt in zip(X, y):
        yh = 1 if w[0]*x_i[0]+w[1]*x_i[1]+b >=0 else 0
        w[0] += lr*(yt - yh)*x_i[0]
        w[1] += lr*(yt - yh)*x_i[1]
        b += lr*(yt - yh)

correct = sum(1 for x_i, yt in zip(X,y) if (1 if w[0]*x_i[0]+w[1]*x_i[1]+b >=0 else 0)==yt)
print("Final w:", w)
print("Final b:", b)
print("Training accuracy:", correct/len(y))
```

```python
sentences = ["genai is fun","genai is useful","ai is useful","genai can write code","ai can write code"]
bigram_counts = {}
for s in sentences:
    words = s.split()
    for i in range(len(words)-1):
        bigram_counts.setdefault(words[i],{})
        bigram_counts[words[i]][words[i+1]] = bigram_counts[words[i]].get(words[i+1],0)+1

def predict_next(word):
    return max(bigram_counts[word], key=bigram_counts[word].get) if word in bigram_counts else None

print("Bigram table:", bigram_counts)
print("Predictions:")
print("genai ->", predict_next("genai"))
print("is ->", predict_next("is"))
print("can ->", predict_next("can"))
```
